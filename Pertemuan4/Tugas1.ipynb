**TUGAS 1 - BFS** 

```python
# Maulana Bintang Irfansyah
graph = {
    'A' : ['B'],
    'B' : ['F', 'G'],
    'C' : ['D'],
    'D' : ['H'],
    'E' : [],
    'F' : ['E'],
    'G' : ['C']
}

def bfs(graph, start, end):
  queue = []
  queue.append([start])
  while queue:
    path = queue.pop(0)
    node = path[-1]
    if node == end:
      return path
    
    for adjacent in graph.get(node, []):
      new_path = list(path)
      new_path.append(adjacent)
      queue.append(new_path)

print('Hasil penelusuran node D pada graf dengan BFS:')
bfs(graph, 'A', 'D')
```
