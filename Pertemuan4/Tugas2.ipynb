**Tugas 2 - DFS**

```python
# Maulana Bintang Irfansyah
graph = {
    'A' : ['B'],
    'B' : ['F', 'G'],
    'C' : ['D'],
    'D' : ['H'],
    'E' : [],
    'F' : ['E'],
    'G' : ['C']
}

def dfs_path(graph, start, end):
  result = []
  dfs(graph, start, end, [], result)
  return result

def dfs(graph, start, end, path, result):
  path+=[start]
  if start == end:
    result.append(path)
  else:
    for node in graph[start]:
      if node not in path:
        dfs(graph, node, end, path[:], result)

print('Hasil penelusuran node D pada graf dengan DFS:')
dfs_path(graph, 'A', 'D')
```
